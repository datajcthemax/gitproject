```python
import pandas as pd

df = pd.read_csv('C:/Users/Playdata/Desktop/찌르레기/gitproject/datas/train.csv')

df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.drop('ID', axis=1, inplace=True)
```

```python
df
```

```python
dummy_vars = pd.get_dummies(df[['Weight_Status','Gender']])
```

```python
df = pd.concat([df, dummy_vars], axis=1)
```

```python
df
```

```python
df.drop(columns=['Weight_Status', 'Gender'], inplace=True)
```

```python
df
```

```python
# create a new column 'Height (inches)' by multiplying 'Height(Feet)' by 12 and adding 'Height(Remainder_Inches)'
df['Height (inches)'] = df['Height(Feet)'] * 12 + df['Height(Remainder_Inches)']

# drop the original columns 'Height(Feet)' and 'Height(Remainder_Inches)'
df.drop(columns=['Height(Feet)', 'Height(Remainder_Inches)'], inplace=True)
```

```python
df
```

```python
df['BMI'] = (df['Weight(lb)'] / (df['Height (inches)'] ** 2)) * 703
```

```python
df
```

```python
df['TBSA'] = 0.20247 * (df['Height (inches)'] ** 0.725) * (df['Weight(lb)'] ** 0.425)
```

```python
df['Body_Fat_Percentage'] = (1.2 * df['BMI']) + (0.23 * df['Age']) - (10.8 * (1 - df['Gender_F'])) - 5.4
```

```python
df['RMR'] = (10 * df['Weight(lb)']) + (6.25 * df['Height (inches)']) - (5 * df['Age']) + 5
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler

# create a scaler object
scaler = MinMaxScaler()

# fit and transform the data, excluding the "Calories_Burned" column
df_scaled = df.drop(columns=['Calories_Burned'])
df_scaled = scaler.fit_transform(df_scaled)

# create a new dataframe with the scaled data
df_scaled = pd.DataFrame(df_scaled, columns=df.columns[:-1])
df_scaled['Calories_Burned'] = df['Calories_Burned']


```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

X = df.drop(['Calories_Burned'], axis=1)
y = df['Calories_Burned']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=100)



```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
y_pred = linreg.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print('Root mean squared error:', rmse)
```
